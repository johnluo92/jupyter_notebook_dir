```python
def try_sudoku(P):

    for row, col in [(r, c) for r in range(9) for c in range(9) if not P[r][c]]:
            
        rr, cc = (row // 3) * 3, (col // 3) * 3
            
        use = {1,2,3,4,5,6,7,8,9} - ({P[row][c] for c in range(9)} | {P[r][col] for r in range(9)} | {P[rr+r][cc+c] for r in range(3) for c in range(3)})

        if len(use) == 1:
            P[row][col] = use.pop()
            return try_sudoku(P)
    return P
puzzle = [[5,3,0,0,7,0,0,0,0],
          [6,0,0,1,9,5,0,0,0],
          [0,9,8,0,0,0,0,6,0],
          [8,0,0,0,6,0,0,0,3],
          [4,0,0,8,0,3,0,0,1],
          [7,0,0,0,2,0,0,0,6],
          [0,6,0,0,0,0,2,8,0],
          [0,0,0,4,1,9,0,0,5],
          [0,0,0,0,8,0,0,7,9]]
try_sudoku(puzzle)
```

```python
def bubble_sort(arr):
    
    def swap(i, j):
        arr[i], arr[j] = arr[j], arr[i]

    n = len(arr)
    swapped = True
    
    x = -1
    while swapped:
        swapped = False
        x = x + 1
        for i in range(1, n-x):
            if arr[i - 1] > arr[i]:
                swap(i - 1, i)
                swapped = True
                    
    return arr

print(bubble_sort([5,2,0,1,4,3]))
```
